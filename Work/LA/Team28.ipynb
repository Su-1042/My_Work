### 1. Degree Centrality
  Consider a graph G := (V,E) with V as the number of vertices and E as the number of edges.
  Centrality of a node v is given as :

  $C_{D} = deg(v)$

  let **v*** be the node in the graph with highest degree centrality .

  consider a subgraph **X** := (Y , Z) which is a Y node connected graph .

   $H=\sum _{{j=1}}^{{|Y|}}[C_{D}(y*)-C_{D}(y_{j})]$

   Calculating the degree Centralization of the graph G .
   $$C_D(G) = \frac{\sum_{i=1}^{|V|} [C_D(v^*) - C_D(v_i)]}{H}$$

   The value of **H** is maximised when **X** is a star graph .

   Screenshot from 2023-06-20 17-13-43.png


  consider a **n** node star graph then the value of H can be computed as :

  ${\displaystyle H=(n-1)\cdot ((n-1)-1)=n^{2}-3n+2.}$

  So for any graph G := ( V,E ) ,

$$C_D(G) = \frac{\sum_{i=1}^{V} [C_D(v^*) - C_D(v_i)]}{|V|^2 - 3|V| + 2}$$









###2. Closeness Centrality
Consider a graph G := ( V , E ) . Closeness Centrality was defined as the reciprocal of farness , Let d(u,v) be distance between nodes u and v belonging to the graph G .
  Then centrality of any node v is defined as follows :


  $$C(v) = \frac{N-1}{\sum_{u}d(u,v)}$$

  from various observations we can get a relationship between degree centrality and Closeness centrality as follows :



  $${\displaystyle (C(v))^{-1}\approx -\alpha \ln(k_{v})+\beta }$$


here α and β are constants and $k_{v}$ is the degree centrality of node v .

### 3.  Betweeness Centrality
  Betweeness Centrality quantifies the number of times a node v behaves a bridge for shortest path between different nodes of the graph .

  Consider a graph G := ( V , E ) with V vertices and E edges .
  For evaluating the closeness centrality first we need to compute shortest path between all the nodes of the graph G using Floyd Warshall in θ( $v^{3}$
 that is constant time complexity .

  Let say we want to calculate the centrality for node v in the graph G .
  We need to calculate the number of shortest paths passing through this vertex v . Then divide it with the total number of shortest path in a graph G.

  Let G be a n node graph then the total number of shortest paths will be
  **${{n}\choose{x}}$** for the undirected graph as same path to reach from u to v and v  to u  . If the graph is a directed graph the total number of shortest paths will be 2 times **${{n}\choose{x}}$** as the paths to reach from u to v is different from v to u .

  So for any vertice v in the graph G centrality can be represented as :
  $$C_B(v) = \sum_{s \neq v \neq t \in V}\frac{\sigma_{st}(v)}{\sigma_{st}}$$
  Where ${\sigma_{st}}$ is the total number of shortest paths in the graph G .







### 4.Eigen Centrality
It is general form of **Degree centrality** , in degree centrality we only account for the number of connections a node has , in Eigen Centrality we also consider the number of connections the neighbours have .

It assigns relative scores to all nodes in the network based on the concept that connections to high-scoring nodes contribute more to the score of the node in question than equal connections to low-scoring nodes.

##CENTRALITIES IN IN SOCIAL NETWORK ANALYSIS

---




Centrality is a fundamental concept in social network analysis that allows us to quantify the importance or influence of an individual within a social network. It provides a way to assess the structural position and prominence of a node (representing an individual) based on its connections and interactions with other nodes in the network.

There are various centrality measures that can be applied to analyze social networks, each capturing a different aspect of importance.A few of them that are used and theor role in an organisational system are described as follows

1. **Degree Centrality**: Degree centrality is the simplest centrality measure and is calculated by counting the number of connections (edges) that a node has. In an organizational context, an individual with high degree centrality would have numerous direct connections or interactions with other members. They are likely to be well-connected, influential, and have access to diverse information and resources. Such individuals often play crucial roles in coordinating and disseminating information within the organization.


2. **Betweenness Centrality**: Betweenness centrality quantifies the extent to which a node lies on the shortest paths between other pairs of nodes. In an organization, a person with high betweenness centrality acts as a bridge or intermediary between different groups or individuals who are not directly connected. They have control over the flow of information and can facilitate communication and collaboration between otherwise disconnected parts of the organization. Individuals with high betweenness centrality often play important brokerage roles.


3. **Eigenvector Centrality**: Eigenvector centrality takes into account both the number of connections and the centrality of those connections. It assigns higher scores to nodes that are connected to other important nodes. In an organizational context, an individual with high eigenvector centrality would have connections to other highly connected and influential individuals. They are considered important not only because of their own connections but also because of the connections they have to other key individuals within the organization.


4. **Closeness Centrality**: Closeness centrality measures how quickly a node can reach all other nodes in the network. In an organization, an individual with high closeness centrality is likely to have direct access to a large portion of the organization. They can disseminate information efficiently and are well-positioned to influence decision-making processes. Such individuals often have a strong presence and can quickly mobilize resources or support.


5. **Katz Centrality**: Katz Centrality extends the concept of eigenvector centrality by considering both direct and indirect connections between nodes in a network. It takes into account the entire network structure and assigns higher centrality scores to nodes that are not only directly connected to other important nodes but also connected to them indirectly.Thus, it captures both direct and indirect influence within a network. It provides a more comprehensive understanding of an individual's centrality and their role in spreading information and influencing others in the network.


These centrality measures provide different perspectives on an individual's importance within a social network.Combining multiple centrality measures can provide a more comprehensive understanding of an individual's centrality and influence within a social network.




```python
import networkx as nx
import matplotlib.pyplot as plt

def generate_connected_graph(num_nodes, avg_degree, rewiring_prob):
    while True:
        graph = nx.connected_watts_strogatz_graph(num_nodes, avg_degree, rewiring_prob)
        if nx.is_connected(graph):
            return graph

num_nodes = 100
avg_degree = 5
rewiring_prob = 0.4

graph = generate_connected_graph(num_nodes, avg_degree, rewiring_prob)

print("Graph Information:")
print(f"Number of nodes: {len(graph.nodes)}")
print(f"Number of edges: {len(graph.edges)}")

pos = nx.spring_layout(graph, k=2.0, scale=1, iterations=200)
nx.draw(graph, pos, with_labels=True, node_color='lightblue', node_size=500, edge_color='gray')
plt.title("Random Graph Visualization")
plt.gcf().set_size_inches(8, 8)
plt.show()
```

```python
degree_centrality = {}

no_nodes = len(graph.nodes)
for node in graph.nodes:
    degree = graph.degree[node]
    degree_centrality[node] = degree/(no_nodes-1)


def closeness_centrality(graph):
    closeness_dict = {}
    n = len(graph.nodes())

    for node in graph.nodes():
        shortest_paths = nx.shortest_path_length(graph, node)
        total_distance = sum(shortest_paths.values())

        if total_distance == 0:
            closeness_centrality = 0
        else:
            reachable_nodes = len(shortest_paths) - 1
            closeness_centrality = reachable_nodes / total_distance

        closeness_dict[node] = closeness_centrality

    return closeness_dict

closeness_centrality = closeness_centrality(graph)
```

```python
eigenvector_centrality = nx.eigenvector_centrality(graph, max_iter = 1000)

betweenness_centrality = nx.betweenness_centrality(graph)

katz_centrality = nx.katz_centrality(graph, alpha=0.1, beta=1.0)
```

```python
def normalize_scores(scores):
    min_score = min(scores.values())
    max_score = max(scores.values())

    normalized_scores = {}
    for tag, score in scores.items():
        normalized_score = (score - min_score) / (max_score - min_score)
        normalized_scores[tag] = normalized_score

    return normalized_scores

degree_centrality = normalize_scores(degree_centrality)
closeness_centrality = normalize_scores(closeness_centrality)
eigenvector_centrality = normalize_scores(eigenvector_centrality)
betweenness_centrality = normalize_scores(betweenness_centrality)
katz_centrality = normalize_scores(katz_centrality)
```

### ANALYSIS

In the given scenario, we are analyzing a network of 100 people with a total of 200 connections. To understand the network's structure and identify the most influential nodes, various centrality measures are calculated individually. Centrality measures provide insights into the importance or influence of a node within a network.

First, let's discuss the different centrality measures that can be calculated. Commonly used centrality measures include degree centrality, betweenness centrality, closeness centrality, and eigenvector centrality. Degree centrality calculates the number of connections each node has. Betweenness centrality identifies nodes that act as crucial bridges between other nodes. Closeness centrality measures the average distance from a node to all other nodes in the network. Eigenvector centrality assesses the influence of a node based on its connections to other influential nodes.

Once these centrality measures are calculated for each node, a weight is assigned to each centrality based on its significance in the network. This weight represents the relative importance of each centrality measure. For example, if betweenness centrality is considered more important than degree centrality, it would be assigned a higher weight.

To ensure that the centrality scores are comparable, each centrality measure is normalized between 0 and 1. Normalization scales the values to a common range, allowing for meaningful comparison and combination of different centrality measures.

To calculate the weighted centrality score for each node, each centrality measure is multiplied by its corresponding weight and added together. This process combines the individual centrality measures, considering their relative importance according to the assigned weights.

Finally, the graph is replotted to visualize the most influential node(s) based on the weighted centrality scores. The nodes with higher centrality scores, reflecting their overall influence in the network, are highlighted using different colors or markers. This visualization helps identify and understand the most significant nodes in the network, as determined by the weighted centrality scores.

By analyzing the network using graph-based calculations and assigning weights to centrality measures, we can gain insights into the influential nodes and their importance in the overall network structure.

```python
import networkx as nx
import random
import matplotlib.pyplot as plt

# Create an network graph with 50 people
def generate_connected_graph(num_nodes, avg_degree, rewiring_prob):
    while True:
        graph = nx.connected_watts_strogatz_graph(num_nodes, avg_degree, rewiring_prob)
        if nx.is_connected(graph):
            return graph

num_nodes = 50
avg_degree = 7
rewiring_prob = 0.4
infection_probability = 0.2
# Here we have taken the probability of the spread of the disease from an infected person to another person as 0.2

G = generate_connected_graph(num_nodes, avg_degree, rewiring_prob)
Gtemp=G.copy()

# Function to simulate the spreading process
def simulate_spreading(graph, infection_probability):
    infected = set()
    infected.add(random.choice(list(graph.nodes())))

    while True:
        new_infected = set()
        for node in infected:
            neighbors = list(graph.neighbors(node))
            for neighbor in neighbors:
                if neighbor not in infected:
                    if random.random() < infection_probability:
                        new_infected.add(neighbor)
        if len(new_infected) == 0:
            break
        infected.update(new_infected)

    return infected

def normalize_scores(scores):
    min_score = min(scores.values())
    max_score = max(scores.values())

    normalized_scores = {}
    for tag, score in scores.items():
        normalized_score = (score - min_score) / (max_score - min_score)
        normalized_scores[tag] = normalized_score

    return normalized_scores

# Calculate the network spread probability
num_simulations = 1000
spread_count = 0

for _ in range(num_simulations):
    infected_nodes = simulate_spreading(G, infection_probability)
    if len(infected_nodes) == num_nodes:
        spread_count += 1

spread_probability = spread_count / num_simulations

degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G)
katz_centrality = nx.katz_centrality(G)

degree_centrality=normalize_scores (degree_centrality)
betweenness_centrality=normalize_scores (betweenness_centrality)
closeness_centrality=normalize_scores (closeness_centrality)
katz_centrality=normalize_scores(katz_centrality)
most_influential_degree = max(degree_centrality, key=degree_centrality.get)
most_influential_betweenness = max(betweenness_centrality, key=betweenness_centrality.get)
most_influential_closeness = max(closeness_centrality, key=closeness_centrality.get)
most_influential_eigenvector = max(eigenvector_centrality, key=eigenvector_centrality.get)
most_influential_katz = max(katz_centrality, key=katz_centrality.get)

print("Most Influential Node based on Degree Centrality:", most_influential_degree)
print("Most Influential Node based on Betweenness Centrality:", most_influential_betweenness)
print("Most Influential Node based on Closeness Centrality:", most_influential_closeness)
print("Most Influential Node based on Eigenvector Centrality:", most_influential_eigenvector)
print("Most Influential Node based on Katz Centrality:", most_influential_katz)
print("Network Spread Probability:", spread_probability)
print("\n")


plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos=pos, with_labels=True, node_color='lightblue', node_size=200)
plt.title("Original Network Graph")
plt.show()


# Now we remove the two nodes with the highest Katz centrality
G.remove_node(most_influential_katz)
katz_centrality.pop(most_influential_katz)
second_most_influential_katz = max(katz_centrality, key=katz_centrality.get)
G.remove_node(second_most_influential_katz)

# Calculating the network spread probability after removal of two nodes with highest Katz centrality
spread_count = 0

for _ in range(num_simulations):
    infected_nodes = simulate_spreading(G, infection_probability)
    if len(infected_nodes) == num_nodes - 2:
        spread_count += 1

spread_probability_after_katz_removal = spread_count / num_simulations

plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos=pos, with_labels=True, node_color='lightblue', node_size=200)
plt.title("Network Graph after Removal of Nodes with Highest Katz Centrality")
plt.show()

# Now we remove the two nodes with the highest Betweenness centrality
G = Gtemp.copy()
betweenness_centrality = nx.betweenness_centrality(G)

most_influential_betweenness = max(betweenness_centrality, key=betweenness_centrality.get)
G.remove_node(most_influential_betweenness)
betweenness_centrality.pop(most_influential_betweenness)
second_most_influential_betweenness = max(betweenness_centrality, key=betweenness_centrality.get)
G.remove_node(second_most_influential_betweenness)

spread_count = 0

for _ in range(num_simulations):
    infected_nodes = simulate_spreading(G, infection_probability)
    if len(infected_nodes) == num_nodes - 2:
        spread_count += 1

spread_probability_after_betweenness_removal = spread_count / num_simulations


# Calculation based on the removal of nodes with the highest degree centrality

# G = Gtemp.copy()  # Regenerate the graph
# degree_centrality = nx.degree_centrality(G)

# most_influential_degree = max(degree_centrality, key=degree_centrality.get)
# G.remove_node(most_influential_degree)
# degree_centrality.pop(most_influential_degree)
# second_most_influential_degree = max(degree_centrality, key=degree_centrality.get)
# G.remove_node(second_most_influential_degree)

# # Calculating the network spread probability after removal of two nodes with highest Betweenness centrality
# spread_count = 0

# for _ in range(num_simulations):
#     infected_nodes = simulate_spreading(G, infection_probability)
#     if len(infected_nodes) == num_nodes - 2:
#         spread_count += 1

# spread_probability_after_degree_removal = spread_count / num_simulations
# print("Network Spread Probability after removing two nodes with highest Betweenness centrality:", spread_probability_after_degree_removal)
# print("\n")


# Plot the graph after removing nodes with highest Katz centrality

# Regenerate the graph and plot after removing nodes with highest Betweenness centrality

# Remove nodes with highest Betweenness centrality

plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos=pos, with_labels=True, node_color='lightblue', node_size=200)
plt.title("Network Graph after Removal of Nodes with Highest Betweenness Centrality")
plt.show()


print("Network Spread Probability after removing two nodes with highest Katz centrality:", spread_probability_after_katz_removal)
print("Network Spread Probability after removing two nodes with highest Betweenness centrality:", spread_probability_after_betweenness_removal)
```

### Analysis of Centralities in Disease Spread:

---

In the given network graph, we have calculated various centrality measures to analyze the spread of a disease. These centrality measures provide insights into the importance of individual nodes in the network and can help us identify influential nodes that play a significant role in disease transmission.

1. Degree Centrality: Degree centrality measures the number of connections a node has. Nodes with higher degree centrality are more connected and have a higher potential to spread the disease. In our analysis, we identified the node with the highest degree centrality as the most influential node based on this measure.

2. Betweenness Centrality: Betweenness centrality quantifies how often a node acts as a bridge in the shortest paths between other nodes. Nodes with higher betweenness centrality can control the flow of information or disease transmission in the network. Identifying nodes with high betweenness centrality helps us pinpoint critical points for intervention and control.

3. Closeness Centrality: Closeness centrality measures how close a node is to other nodes in terms of shortest path distances. Nodes with higher closeness centrality are more accessible to other nodes and can spread the disease more efficiently.

4. Eigenvector Centrality: Eigenvector centrality considers both the node's connections and the importance of its neighbors. Nodes with higher eigenvector centrality are well-connected to other influential nodes and have a higher potential to spread the disease.

5. Katz Centrality: Katz centrality combines both local and global information by considering the number of immediate neighbors and their respective centralities. It captures the influence of nodes based on both direct and indirect connections. **In our analysis, Katz centrality was identified as the most important centrality measure.**

The Importance of Katz Centrality: Katz centrality provides a comprehensive measure of node influence, considering both direct and indirect connections. In the context of disease spread, nodes with high Katz centrality are crucial in controlling and containing the spread. These nodes have a strong influence on neighboring nodes, including those with high degrees or betweenness centralities. Identifying nodes with high Katz centrality allows us to focus on strategic interventions to mitigate the spread effectively.

We have calculated the Network Spread Probability, i.e the probability of the spread of the disease in the given network.
